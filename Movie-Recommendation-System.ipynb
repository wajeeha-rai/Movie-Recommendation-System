```python
def get_name(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name']) 
    return L 
```

```python
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')

def preprocess_text(text):
    tokens = text.lower().split()
    filtered_tokens = [word for word in tokens if word not in stopwords.words('english')]
    return " ".join(filtered_tokens)
```

```python
merge_df['tags']=merge_df['tags'].apply(preprocess_text)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer(stop_words='english', ngram_range=(1, 2), max_df=0.85, min_df=2)
x = vectorizer.fit_transform(merge_df['tags']).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cos_sim = cosine_similarity(x)
```

```python
cos_sim
```

```python
def recommend_movies(movie_name, merge_df, cosine_sim_matrix, top_n=10):
    if movie_name not in merge_df['title_x'].values:
        return f"Movie '{movie_name}' not found in the dataset."
        
    movie_index = merge_df[merge_df['title_x'] == movie_name].index[0]
    similarity_scores = cosine_sim_matrix[movie_index]
    similar_movies_indices = similarity_scores.argsort()[-(top_n + 1):-1]
    movie_recommendations = []
    for i in similar_movies_indices:
        movie_recommendations.append((merge_df['title_x'][i], similarity_scores[i]))
    
    return movie_recommendations
```

```python
movie_name = input("Enter movie title: ")
recommended_movies = recommend_movies(movie_name, merge_df, cos_sim, top_n=10)
for i, (title, _) in enumerate(recommended_movies, 1):
    print(f"{i}. {title}")
```
